```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Region': ['North', 'South', 'East', 'West', 'North'],
    'Product': ['Laptop', 'Monitor', 'Keyboard', 'Mouse', 'Monitor'],
    'Sales': [25000, 12000, 5000, 3000, 10000]
}

df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)

# Read the CSV file again
df = pd.read_csv('sales_data.csv')
df.head()

```

```python
region_sales = df.groupby('Region')['Sales'].sum()
print("Total Sales by Region:\n", region_sales)

region_sales.plot(kind='bar', title='Total Sales by Region', color='skyblue')
plt.ylabel('Sales')
plt.xlabel('Region')
plt.grid()
plt.show()
```

```python
# Total sales by Product
product_sales = df.groupby('Product')['Sales'].sum()
print("Total Sales by Product:\n", product_sales)

# Plot a horizontal bar chart
product_sales.plot(kind='barh', title='Total Sales by Product', color='orange')
plt.xlabel('Sales')
plt.ylabel('Product')
plt.grid()
plt.show()
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group by month and sum sales
monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Sales'].sum()

# Plot the line chart
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.grid()
plt.show()
```
